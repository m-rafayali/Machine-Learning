```python
# lets visualize using 2d in pca
```

```python
df=pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaleddata=scaler.transform(df)
```

# PCA

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
# inhi 2 components se data visualize krenge
```

```python
pca.fit(scaleddata)
```

```python
x_pca=pca.transform(scaleddata)
```

```python
scaleddata.shape
```

```python
x_pca.shape
# 30 se 2 reducce hogyi
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1])
# from col0 to plot against col1
plt.xlabel('1st pca')
plt.ylabel('2nd pca')
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1],c=cancer['target'])
# from col0 to plot against col1
plt.xlabel('1st pca')
plt.ylabel('2nd pca')

# color these cols acc to malignat belign i.e target
```

componemts corresponds to combinations of original features
array m each row represnt principle component
each cols actually represnts original features

lets visualize this relationship with heatmap

```python
pca.components_
```

```python
df_comp=pd.DataFrame(pca.components_,columns=cancer['feature_names'])
```

```python
df_comp
# pca 1 pca 2 ke features agaye i.e 2 rows
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(df_comp)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
y=cancer['target']
```

```python
lm.fit(x_pca,y)
```

```python
print(lm.intercept_)
```

```python
print(lm.coef_)
```

```python
predictions=lm.predict(x_pca)
predictions
```

```python
# Use only the first principal component
plt.scatter(x_pca[:, 0], y, color='blue', label='Actual values')

# Plot the regression line based on the first component
plt.plot(x_pca[:, 0], predictions, color='red', label='Regression line')

plt.xlabel('PCA Component 1')
plt.ylabel('Target Variable')
plt.title('Linear Regression on PCA Component')
plt.legend()
plt.show()
```
