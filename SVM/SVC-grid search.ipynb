```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer=load_breast_cancer()
```

```python
cancer.keys()
```

```python
print(cancer['DESCR'])
# Description of data 
```

```python
# to predict whether tumor is mailgnant or belign
```

```python
dffeat=pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
```

```python
dffeat.head()
```

```python
cancer['target_names']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=dffeat
y=cancer['target']
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=101
)
```

```python
from sklearn.svm import SVC
```

```python
model=SVC()
```

```python
model.fit(X_train,y_train)
```

```python
pred=model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(confusion_matrix(y_test,pred))
print('\n')
print(classification_report(y_test,pred))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={'C':[0.1,1,10,100,1000],'gamma':[1,0.1,0.01,0.001,0.0001]}
```

```python
grid=GridSearchCV(SVC(),param_grid,verbose=3)
# gridsearchcv is a meta estimator -- it takes a estimator like we did with SVC and craete a new estimator
# and that works exactly the same
# the higher the num the more verbose its just the text description of output
```

```python
grid.fit(X_train,y_train)
# ye va;ues verbose output krra
```

```python
# isne khudhi testing krli parameters ki ayur jo best parameter tha usko choose bhi krliya
# ab hum best parameters print krkr dekhte
grid.best_params_
# ye best parameter hain
```

```python
grid.best_estimator_
```

```python
# ab is best parameters pe predictions run krrk dekhte hain
grid_pred=grid.predict(X_test)
```

```python
print(confusion_matrix(y_test,grid_pred))
print('\n')
print(classification_report(y_test,grid_pred))
```
