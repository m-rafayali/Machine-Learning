```python

# Step 1: Calculate the average employment length for each title
avg_emp_length_per_title = df.groupby('emp_title')['emp_length'].mean().reset_index()

# Rename the columns for clarity
avg_emp_length_per_title.columns = ['emp_title', 'avg_emp_length']

# Step 2: Create a mapping from emp_title to average emp_length
emp_length_mapping = dict(zip(avg_emp_length_per_title['emp_title'], avg_emp_length_per_title['avg_emp_length']))

# Step 3: Fill missing emp_length values
df['emp_length'] = df['emp_length'].fillna(df['emp_title'].map(emp_length_mapping))

# Step 4: Create a new mapping for filling emp_title based on emp_length
# Here, we want to find out which title corresponds to each emp_length
# We need to ensure we only use unique values for mapping
unique_emp_length_mapping = avg_emp_length_per_title.drop_duplicates(subset=['avg_emp_length'])

# Create a dictionary from the unique average emp_length to emp_title
length_to_title_mapping = dict(zip(unique_emp_length_mapping['avg_emp_length'], unique_emp_length_mapping['emp_title']))

# Fill missing emp_title values based on the emp_length
df['emp_title'] = df['emp_title'].fillna(df['emp_length'].map(length_to_title_mapping))

# Display the updated DataFrame
print(df[['emp_title', 'emp_length']])
```

```python
X=df.drop('loan_status_numeric',axis=1).values
y=df['loan_status_numeric'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
...     X, y, test_size=0.25, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop=EarlyStopping(monitor='val_loss',mode='min',
                        verbose=1,patience=25)
model=Sequential()
# binary hai jbhi end layer m sigmoid add krre
model.add(Dense(78,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(39,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(19,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam')

```

```python
model.fit(x=X_train,y=y_train,epochs=600,validation_data=(X_test,y_test),
         callbacks=[early_stop])
```

```python
losses=model.history.history
```

```python
loss=pd.DataFrame(losses)
loss
```

```python
loss.plot()
```

```python
# Get the predicted probabilities (between 0 and 1)
predictions = model.predict(X_test)

# Convert probabilities to binary class labels (0 or 1)
predicted_classes = (predictions > 0.5).astype(int)

print(predicted_classes.flatten())  # Flatten to print class labels like [1 1 1 0 0 1 1]
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,predicted_classes))
print('\n')
print(classification_report(y_test,predicted_classes))
```
